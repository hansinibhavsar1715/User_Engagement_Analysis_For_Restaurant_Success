```python
import pandas as pd

chunks = pd.read_json(
    r"C:\Users\Dhanshri\Desktop\Yelp_Project\yelp_academic_dataset_review.json",
    lines=True,
    chunksize=50000
)

final_df = pd.DataFrame()

for chunk in chunks:
    chunk = chunk[['review_id','business_id','stars','useful','funny','cool','date']]
    final_df = pd.concat([final_df, chunk])
    
    if len(final_df) >= 200000:
        break

final_df = final_df.sample(200000)
final_df.to_csv("review_small.csv", index=False)
```
